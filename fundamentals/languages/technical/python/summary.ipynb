## **UNIT 0** - _HOW IT STARTED_

### **➔ PYTHON HISTORY**

#### • Who is the Creator

> **Python** created by the Dutch programmer **Guido van Rossum** in the late 1980s and **first released** in **1991**. He developed the language at the **Centrum Wiskunde & Informatica (CWI)** in the **Netherlands** and led its development for many years, being known as the **“Benevolent Dictator For Life”** of the project.

#### • Why Python Was Created

> Guido van Rossum designed Python to be **a language that was easy to read and write**, correcting limitations he saw in earlier languages like ABC. His goal was to **create a general-purpose language with clear syntax, strong readability, and features that would reduce the effort needed to write and maintain code**.

#### • Objective and Uses of Python

> Python’s objective is to **be a versatile, general-purpose language that can be used across many fields**. Because of its simplicity and large ecosystem of libraries, Python is widely used in areas such as:
> - **web development**
> - **data science**
> - **machine learning**
> - **automation**
> - **scientific computing**
> - **artificial intelligence**

### **➔ PYTHON PARTICULAR THINGS**

Python’s elegant syntax and dynamic typing, together with its interpreted nature, make it an ideal language for scripting and rapid application development in many areas on most platforms.


#### • Free and Open Source

> Python is **free** to use and its source code is publicly available. This means **anyone** can **study**, **modify**, and **share it**. In practice, this creates a **large community** and **many libraries**, making development easier and more accessible.

#### • High-level language

> That means it is **easy to learn and use** because it is closer to the **human language** than to Assembly language.

#### • Interpreted

> Python is an **interpreted language**, meaning the code **is executed at runtime** instead of being fully compiled beforehand. In practice, you can **run programs immediately without a compilation step**, which makes **development faster**.

#### • Multi-paradigm

> That means Python can be used with different paradigms, such as:
> - **Procedural programming**
> - **Object-oriented programming**
> - **Functional programming**

#### • Dynamically and Strongly Typed

> Python is dynamically typed, meaning **variable types are determined at runtime** and **do not need to be declared**. It is also strongly typed, meaning **different types are not automatically converted**. In practice, you must **manually convert types** when necessary.

#### • Plataforms used

> Python is **cross-platform**, meaning it **runs on different operating systems**. In practice, **the same code** can **work** on Windows, macOS, and Linux **with little or no modification**.

### **➔ COMMENTS TYPES**

#### • Type 1 - Inline Comments

```python
a = 10 # This variable stores the value 10 and this is an exemple of inline comment
print(a)
```

#### • Type 2 - Comments in block

```python
a = 10 
# This is an example of block comment:
# In Python, don't exist a especifically syntax for block comments
# but we can use multiple lines of comments to do the same thing.
print(a)
```

#### • Type 3 - Comments for documentation (docStrings)

> In Python, we can use docstrings to describe the purpose of a function, class, or other types of objects. It looks like a block comment, but it is actually a string that can be accessed using the __doc__ attribute and is displayed when using the help() function.
>
> In this example, we use a docstring to describe the purpose of the function showObject:

```python
def showObject(object):
    """
    This function receives an object and prints it.
    """
    print(object)

a = "Hello, World!"
showObject(a)

help(showObject)
```
